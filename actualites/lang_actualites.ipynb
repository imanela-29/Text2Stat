```python
f = open('actualites.txt', 'r', encoding='utf-8')

fr_output = open('fr-actualites.txt', 'a', encoding='utf-8')
ar_output = open('ar-actualites.txt', 'a', encoding='utf-8')
```

```python
for x in f:
    # Language Detection
    try:
        lang2 = detect(x)
        if lang2 == 'ar':
            ar_output.write(x)
            ar_output.write("\n")
        else:
            fr_output.write(x)
            fr_output.write("\n")
    except:
        fr_output.write(x)
        fr_output.write("\n")
```

```python
fr_output.close()
ar_output.close()
f.close()
```
